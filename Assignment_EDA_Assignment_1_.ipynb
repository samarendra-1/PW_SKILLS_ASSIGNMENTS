```python
#   What is the percentage of male and female patients in the dataset?

gender_counts = df['sex'].value_counts()
male_percentage = (gender_counts[1] / df.shape[0]) * 100
female_percentage = (gender_counts[0] / df.shape[0]) * 100
print("Male Percentage:", male_percentage)
print("Female Percentage:", female_percentage)
```

```python
# how does the platlet count vary among different age groups

import matplotlib.pyplot as plt

# Group the data by age and calculate the mean platelet count for each group
age_groups = df.groupby('age')['platelets'].mean()

# Plot the mean platelet count for each age group
age_groups.plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Mean Platelet Count')
plt.title('Variation in Platelet Count by Age')
plt.show()
```

```python
#  Is there is  a correlation between creatinine and sodium levels in the blood

import pandas as pd
import matplotlib.pyplot as plt

# Create a scatter plot of creatinine vs. sodium levels
plt.scatter(df['creatinine_phosphokinase'], df['serum_sodium'])

# Set the title and axis labels
plt.title('Creatinine vs. Sodium Levels')
plt.xlabel('Creatinine Phosphokinase (U/L)')
plt.ylabel('Serum Sodium (mEq/L)')

# Display the plot
plt.show()

# Calculate the correlation coefficient
correlation = df['creatinine_phosphokinase'].corr(df['serum_sodium'])
```

```python
#  how does the prevalence of high blood pressure differ between male and female patients

df.groupby('sex')['high_blood_pressure'].mean()
```

```python
#  what is the relationship between smoking habits and the occurrence of heart failure

import pandas as pd
import seaborn as sns

# Create a cross-tabulation of smoking habits and heart failure occurrence
crosstab = pd.crosstab(df['smoking'], df['DEATH_EVENT'])

# Calculate the percentage of heart failure occurrences for each smoking habit
crosstab['percent'] = crosstab[1] / (crosstab[0] + crosstab[1]) * 100

# Plot the cross-tabulation as a bar chart
sns.barplot(data=crosstab, x='smoking', y='percent')

plt.title ('coz of smoking  heart failure')
plt.show()
```

```python
# prompt: Using dataframe df: Are there any noticaable patterns in the distribution of death events across different age groups

import matplotlib.pyplot as plt

# Create a histogram of the age distribution
plt.hist(df['age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.title('Age Distribution of Patients')
plt.show()

# Create a bar chart of the death event rate by age group
death_rates = df.groupby('age')['DEATH_EVENT'].mean()
plt.bar(death_rates.index, death_rates.values)
plt.xlabel('Age')
plt.ylabel('Death Event Rate')
plt.title('Death Distribution of Patients' )
plt.show()
```

```python
# is there any significant difference in ejection  fraction between patients with and without diabities

import pandas as pd
import scipy.stats as stats

# Group the data by diabetes status
df_diabetes = df[df['diabetes'] == 1]
df_no_diabetes = df[df['diabetes'] == 0]

# Perform a t-test to compare the ejection fractions
t_statistic, p_value = stats.ttest_ind(df_diabetes['ejection_fraction'], df_no_diabetes['ejection_fraction'])

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```
